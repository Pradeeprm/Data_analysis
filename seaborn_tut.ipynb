```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",palette=["r", "b"],sizes=(15,200),data=tips)
```

Titanic CountPlot

```python
titanic = sns.load_dataset("titanic")
titanic[titanic['alive']=="yes"].count()
```

```python
titanic
```

```python
ax = sns.countplot(x="alive", data=titanic)
```

```python
ax = sns.countplot(x="class",hue="sex",data=titanic)
```

```python
sns.barplot(x="sex",y="fare",data=titanic)
```

```python
fig, ax = plt.subplots(1,2)
flights = sns.load_dataset("flights")

apr_flights = flights[flights["month"]=="Apr"]
jan_flights = flights[flights["month"]=="Jan"]

sns.lineplot(x="year",y="passengers",data=apr_flights,ax=ax[0])
sns.countplot(y="passengers",data=jan_flights,ax=ax[1])
fig.suptitle("Apr vs Jan")
fig.show()
```
